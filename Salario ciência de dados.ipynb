# Salário de Ciência de dados

Objetivo de analisar o salario do cientista de dados

[https://www.kaggle.com/datasets/harishkumardatalab/data-science-salary-2021-to-2023](https://www.kaggle.com/datasets/harishkumardatalab/data-science-salary-2021-to-2023)

```python
import pandas as pd
import plotly.express as px
```

```python
base = pd.read_csv('salario.csv')
base.head(3)
```

```python
base['experience_level'].value_counts()
```

```python
base['employment_type'].value_counts()
```

```python
base['experience_level'] = base['experience_level'].replace({'SE': 'Senior', 'MI': 'Nivel Médio', 'EN': 'Nível Básico', 'EX': 'Experiente'})
```

```python
base['employment_type'] = base['employment_type'].replace({'FT': 'Tempo integral', 'PT': 'Tempo Parcial', 'CT': 'Contratado', 'FL': 'Free Lancer'})
```

```python
base.describe()
```

```python
base.info()
```

```python
base.isnull().sum()
```

```python
base['work_year'].value_counts()
```

```python
fig = px.bar(base, x='experience_level', y='salary_in_usd', color='salary_in_usd')
fig.show()
```

```python
fig = px.bar(base, x='employment_type', y='salary_in_usd', color='salary_in_usd')
fig.show()
```

```python
fig = px.histogram(base, x='experience_level', color='employment_type')
fig.show()
```

```python
fig = px.histogram(base, x='employment_type', color='experience_level')
fig.show()
```

```python
base[['job_title', 'salary_in_usd']].groupby(['job_title']).mean().sort_values(by='salary_in_usd', ascending=False).head(20)
```

```python
base['company_location'].value_counts().sort_values(ascending=False).head(20)
```

```python
base.loc[base['company_location'] == 'US', ['job_title', 'salary']].groupby('job_title').mean().sort_values(by='salary', ascending=False)
```

```python
base.loc[base['company_location'] == 'GB', ['job_title', 'salary']].groupby('job_title').mean().sort_values(by='salary', ascending=False)
```

```python
base['company_size'].value_counts()
```

```python
fig = px.bar(base, x='company_size', y='salary_in_usd', color='salary_in_usd', text_auto=True)
fig.show()
```

```python
fig = px.histogram(base, x='company_size', color='experience_level')
fig.show()
```

## Resultados das análise

* Senior Acaba tendo salário mais alto
* profissionais que trabalham em tempo integral tambem tem sálario mais alto
* a maioria dos senior trabalham em tempo integral
* Os cargos mais bem pagos são Lider Tecnico em ciência de dados, arquiteto de dados na nuvem
* para moeda USD o salário mais bem pago são Lider de análise de dados, Lider técnico de analise de dados
* para moeda EUR o salário mais bem pago são Cientista de dados principais, Pesquisador de aprendizagem aprofundada
* para moeda GBP o salário mais bem pago são Arquiteto de big data, Chefe de ciência de dados e pesquisa ciêntifica
* Nos Estados Unidos o salário mais bem pago são de Analista de dados bi, Lider de análise de dados, Lider técnico de ciência de dados
* Na Grã-Bretanha o salário mais bem pago são Análista de dados de marketing, Arquiteto de big data, Chefe de ciência de dados, Arquiteto de dados
* No Canada o salário bem pago são desenvolvedor de IA, Cientista de aprendizado de máquina, Engenheiro de big data, Diretor de ciência de dados
* Nas Média empresas acabam pagando melhor
* O senior são mais requisitado em grandes e média empresa
* O nível básico em pequenas empresa
* e a maioria em tempo integral
